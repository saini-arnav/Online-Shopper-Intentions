## Splitting data and applying ML Algo

```python
x=df.drop('Revenue',axis=1)
x.head()
```

```python
y=df['Revenue']
y.head()
```

```python
from sklearn.model_selection import train_test_split
xtrain,xtest,ytrain,ytest=train_test_split(x,y,test_size=0.25,random_state=101)
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier,ExtraTreeClassifier
from sklearn.ensemble import RandomForestClassifier,AdaBoostClassifier,GradientBoostingClassifier
from sklearn.svm import SVC 
from sklearn.neighbors import KNeighborsClassifier
from sklearn.naive_bayes import GaussianNB
from sklearn.naive_bayes import BernoulliNB


from sklearn.metrics import confusion_matrix, classification_report, accuracy_score 
from sklearn.metrics import f1_score, precision_score, recall_score ,auc,roc_curve   
```

```python
model1=LogisticRegression()
model2=DecisionTreeClassifier()
model3=ExtraTreeClassifier()
model4=RandomForestClassifier()
model5=AdaBoostClassifier()
model6=GradientBoostingClassifier()
model7= SVC() # kernel = linear,poly, sigmoid, rbf
model8= KNeighborsClassifier()
model9= GaussianNB()
model10= BernoulliNB()
```

```python
l=[model1,model2,model3,model4,model5,model6,model7,model8,model9,model10]


def models(x):
    for i in l:
        i.fit(xtrain, ytrain)    
        y_pred_train = i.predict(xtrain)
        y_pred_test = i.predict(xtest)
        print(f"Result for {i}:")
        print("Training Accuracy :", accuracy_score(ytrain, y_pred_train))
        print("Test Accuracy :", accuracy_score(ytest, y_pred_test))
        print('#######'*10)
```

```python
models(l)
```

### Random Forest classifier and Gradient boosting classifier is giving us the best accuracy. 
### let's try some other evaluation matrix to evaluate the model further. 

```python
sns.heatmap(confusion_matrix(ytest,model4.predict(xtest)),annot=True,fmt='0.2f')
plt.savefig('conf1.png')
```

```python
print(classification_report(ytest,model4.predict(xtest)))
# we are getting good precision score for class 0 but not for class 1 due to imbalance
```

### Beacuse of class imbalance , accuracy score alone cannot be a good way to evaluate model so we got with all these:
# Precision
# Recall
# F1-score

```python
yPred=model4.predict_proba(xtest)[:,1]
fpr,tpr,_=roc_curve( ytest, yPred )
area=auc(fpr,tpr)
```

```python
area
```

### PLOTTING ROC CURVE TO GET THE AREA UNDER THE CURVE

```python
plt.figure(figsize=(12,8))
plt.plot(fpr,tpr,color='g',label=f'AUC:{round(area,2)}')
plt.plot([0,1],[0,1],color='r',linestyle='--')
plt.title("ROC CURVE")
plt.xlabel("FPR(FALSE POSITIVE RATE)")
plt.ylabel("TPR(TRUE POSITIVE RATE)")
plt.legend()
plt.savefig('auc1.png')
plt.show()
```

## Using RandomOverSampler from imblearn

## Spliting data and appyling ML algo again

```python
from sklearn.model_selection import train_test_split
xtrain,xtest,ytrain,ytest=train_test_split(x_new,y_new,test_size=0.25,random_state=101)
```

```python
model1=LogisticRegression()
model2=DecisionTreeClassifier()
model3=ExtraTreeClassifier()
model4=RandomForestClassifier()
model5=AdaBoostClassifier()
model6=GradientBoostingClassifier()
model7= SVC() # kernel = linear,poly, sigmoid, rbf
model8= KNeighborsClassifier()
model9= GaussianNB()
model10= BernoulliNB()
```

```python
l=[model1,model2,model3,model4,model5,model6,model7,model8,model9,model10]


def models(x):
    for i in l:
        i.fit(xtrain, ytrain)    
        y_pred_train = i.predict(xtrain)
        y_pred_test = i.predict(xtest)
        print(f"Result for {i}:")
        print("Training Accuracy :", accuracy_score(ytrain, y_pred_train))
        print("Test Accuracy :", accuracy_score(ytest, y_pred_test))
        print('#######'*10)
```

```python
models(l)
```
